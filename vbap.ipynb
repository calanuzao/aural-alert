```python
# import module
import vbap
```

```python
# Choose Azimuth and Elevation in Degrees
az = 40
el = 30
gains = vbap.calculate_vbap(az, el)

print("Speaker Gains (Linear & dB):")
for spk, g in sorted(gains.items(), key=lambda i: -i[1]):
    print(f"{spk}: {g:.3f} ({vbap.lin2db(g):.1f} dB)")
```

```python
import matplotlib.pyplot as plt
import numpy as np

# user inputs
az = 45
el = 30
gains = vbap.calculate_vbap(az, el)

# bar chart of the gains
speakers = list(gains.keys())
gain_values = [gains[s] for s in speakers]
db_values = [vbap.lin2db(g) for g in gain_values]

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 5))

# linear gains
ax1.bar(speakers, gain_values)
ax1.set_title('Speaker Gains (Linear)')
ax1.set_ylim(0, max(gain_values) * 1.1)

# dB gains
ax2.bar(speakers, db_values)
ax2.set_title('Speaker Gains (dB)')
ax2.set_ylim(min([x for x in db_values if x > -30]) - 5, 0)

plt.tight_layout()
plt.show()
```
